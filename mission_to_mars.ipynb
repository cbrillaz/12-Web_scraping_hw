```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs
import requests
import pymongo
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=True)
browser2 = Browser('chrome', **executable_path, headless=True)
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
mars_db = client.mars_db
collection = mars_db.articles
```

```python
url = "https://mars.nasa.gov/news/"
browser.visit(url)

html = browser.html
soup = bs(html, 'html.parser')
    
title = soup.find_all('div', class_='content_title')
lede = soup.find_all('div', class_='article_teaser_body')
```

```python
latest_title = (title[0])
latest_lede = (lede[0])
```

```python
news_title=(latest_title.text)
news_p=(latest_lede.text)

print(news_title)
print(news_p)
```

```python
url2 = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
browser2.visit(url2)

html2 = browser2.html
soup2 = bs(html2, 'html.parser')
```

```python
results = soup2.find_all('li')
print(results)
```
